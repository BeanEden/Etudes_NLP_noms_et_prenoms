# Évaluation Détaillée des Noms de Famille

Ce notebook présente une analyse des performances du modèle de regroupement des noms.

```python
import os
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import evaluation_metrics as em

# Configuration visuelle
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Métriques de Performance

On évalue ici la capacité du modèle à regrouper les variantes orthographiques des noms sans créer de liens erronés.

```python
eval_data = em.load_evaluation_data('./data/2_evaluation_sample.json')
metrics = em.compute_clustering_metrics(eval_data)

print("--- RÉSULTATS NOMS ---")
print(f"Précision : {metrics['precision']:.2%}")
print(f"Rappel    : {metrics['recall']:.2%}")
print(f"F1-Score  : {metrics['f1']:.2%}")

em.plot_confusion_matrix(metrics['confusion_matrix'], title='Matrice de Confusion - Noms de Famille')
```

## 2. Degré de Précision d'après le Modèle (Confidence Score)

Ce score représente la cohésion interne moyenne des groupes. Plus il est proche de 1.0, plus le modèle est "confiant" dans ses regroupements.

```python
conf_score = em.compute_confidence_scores('./data/2_groupes_noms.json', './data/1_noms_clean.json')
print(f"Degré de précision interne moyen (Modèle) : {conf_score:.2f} / 1.00")
```

## 3. Cohérence Étymologique / Géographique

Mesure du pourcentage de groupes dont les membres partagent une même origine déclarée.

```python
ratio, count = em.compute_language_consistency('./data/2_groupes_noms.json', './data/1_noms_clean.json')
print(f"Taux de cohérence d\'origine : {ratio:.2%} (sur {count} groupes multi-membres)")
```
